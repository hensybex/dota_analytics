```python
import requests
import pandas as pd
#import fsspec
from datetime import datetime, timedelta
from io import StringIO
import time
import numpy as np
import time
from ast import literal_eval
```

```python
leaguesDf = pd.read_csv('./leagues.csv')
```

```python
leaguesDf['matches'] = leaguesDf['matches'].replace({'[]': None})
```

```python
leaguesDf = leaguesDf.dropna().reset_index()
```

```python
def convert_str_to_dict(str):
    python_dict = literal_eval(str)
    return python_dict
```

```python
def get_league_id(league_dict):
    return league_dict['leagueid']

def get_league_ticket(league_dict):
    return league_dict['ticket']

def get_league_banner(league_dict):
    return league_dict['banner']

def get_league_tier(league_dict):
    return league_dict['tier']

def get_league_name(league_dict):
    return league_dict['name']
```

```python
leaguesDf['league_info'] = leaguesDf['0'].apply(convert_str_to_dict)
```

```python
leaguesDf = leaguesDf.drop(leaguesDf.columns[[0, 1, 2]], axis=1)
```

```python
leaguesDf['league_id'] = leaguesDf['league_info'].apply(get_league_id)
leaguesDf['league_ticket'] = leaguesDf['league_info'].apply(get_league_ticket)
leaguesDf['league_banner'] = leaguesDf['league_info'].apply(get_league_banner)
leaguesDf['league_tier'] = leaguesDf['league_info'].apply(get_league_tier)
leaguesDf['league_name'] = leaguesDf['league_info'].apply(get_league_name)
```

```python
leaguesDf = leaguesDf.drop(columns=['league_info'])
```

```python
def get_num_of_games(arr):
    return len(arr)
```

```python
leaguesDf['games_num'] = leaguesDf['matches'].apply(get_num_of_games)
```

```python
def matches_to_list(matches):
    matches = matches[1:-1]
    li = list(matches.split(", "))
    return li
```

```python
leaguesDf['matches'] = leaguesDf['matches'].apply(matches_to_list)
```

```python
leaguesDf.to_csv('leagues_parsed.csv')
```
