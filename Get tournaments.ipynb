```python
start = time.time()
for index, row in leaguesDf.iterrows():
    league_id = row[0]['leagueid']
    request_url = 'https://api.opendota.com/api/leagues/' + str(league_id) + '/matches'
    r = requests.get(request_url).text
    matchDf = pd.read_json(r, lines=True).T
    if matchDf.shape[0] != 0:
        tmpMatches = []
        for index2, row2 in matchDf.iterrows():
            tmpMatches.append(row2[0]['match_id'])
        leaguesDf.at[index, 'matches'] = tmpMatches
    else:
        time.sleep(1)
    if index % 100 == 0:
        end = time.time()
        print(index, end - start, "seconds passed")
```

```python
leaguesDf.to_csv('./leagues.csv')
```
