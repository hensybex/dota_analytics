```python
import pandas as pd
from datetime import datetime
import json
import time
```

```python
df = pd.read_csv('./gold_data/full_first_time.csv')
```

```python
def convert_start_time(timestamp):
    return datetime.utcfromtimestamp(timestamp)
```

```python
df['start_time'] = df['start_time'].apply(convert_start_time)
```

```python
dfNew = df.loc[df['start_time'] > timestamp]
```

```python
def fix_json(bad_json):
    return bad_json.replace('\'', '\"')
```

```python
dfNew['player_0'] = dfNew['player_0'].apply(fix_json)
dfNew['player_1'] = dfNew['player_1'].apply(fix_json)
dfNew['player_2'] = dfNew['player_2'].apply(fix_json)
dfNew['player_3'] = dfNew['player_3'].apply(fix_json)
dfNew['player_4'] = dfNew['player_4'].apply(fix_json)
dfNew['player_5'] = dfNew['player_5'].apply(fix_json)
dfNew['player_6'] = dfNew['player_6'].apply(fix_json)
dfNew['player_7'] = dfNew['player_7'].apply(fix_json)
dfNew['player_8'] = dfNew['player_8'].apply(fix_json)
dfNew['player_9'] = dfNew['player_9'].apply(fix_json)
```

```python
#get all unique player_id
all_ids = []
for index, row in dfNew.iterrows():
    all_ids.append(json.loads(row['player_0'])['account_id'])
    all_ids.append(json.loads(row['player_1'])['account_id'])
    all_ids.append(json.loads(row['player_2'])['account_id'])
    all_ids.append(json.loads(row['player_3'])['account_id'])
    all_ids.append(json.loads(row['player_4'])['account_id'])
    all_ids.append(json.loads(row['player_5'])['account_id'])
    all_ids.append(json.loads(row['player_6'])['account_id'])
    all_ids.append(json.loads(row['player_7'])['account_id'])
    all_ids.append(json.loads(row['player_8'])['account_id'])
    all_ids.append(json.loads(row['player_9'])['account_id'])
unique_ids = list(set(all_ids))
```

```python
#mark positions in dfNew
def get_net_worth(player):
    return json.loads(player)['net_worth']

dfNew['player_0_net_worth'] = dfNew['player_0'].apply(get_net_worth)
dfNew['player_1_net_worth'] = dfNew['player_1'].apply(get_net_worth)
dfNew['player_2_net_worth'] = dfNew['player_2'].apply(get_net_worth)
dfNew['player_3_net_worth'] = dfNew['player_3'].apply(get_net_worth)
dfNew['player_4_net_worth'] = dfNew['player_4'].apply(get_net_worth)
dfNew['player_5_net_worth'] = dfNew['player_5'].apply(get_net_worth)
dfNew['player_6_net_worth'] = dfNew['player_6'].apply(get_net_worth)
dfNew['player_7_net_worth'] = dfNew['player_7'].apply(get_net_worth)
dfNew['player_8_net_worth'] = dfNew['player_8'].apply(get_net_worth)
dfNew['player_9_net_worth'] = dfNew['player_9'].apply(get_net_worth)
```

```python
def get_supposed_position_1(match, pos):
    all_nws = []
    all_nws.append(match['player_0_net_worth'])
    all_nws.append(match['player_1_net_worth'])
    all_nws.append(match['player_2_net_worth'])
    all_nws.append(match['player_3_net_worth'])
    all_nws.append(match['player_4_net_worth'])
    all_nws_tmp = all_nws.copy()
    all_nws.sort()
    all_nws_sorted = all_nws.copy()
    if all_nws_tmp[pos] == all_nws_sorted[0]:
        return 5
    elif all_nws_tmp[pos] == all_nws_sorted[1]:
        return 4
    elif all_nws_tmp[pos] == all_nws_sorted[2]:
        return 3
    elif all_nws_tmp[pos] == all_nws_sorted[3]:
        return 2
    else:
        return 1

def get_supposed_position_2(match, pos):
    all_nws = []
    all_nws.append(match['player_5_net_worth'])
    all_nws.append(match['player_6_net_worth'])
    all_nws.append(match['player_7_net_worth'])
    all_nws.append(match['player_8_net_worth'])
    all_nws.append(match['player_9_net_worth'])
    all_nws_tmp = all_nws.copy()
    all_nws.sort()
    all_nws_sorted = all_nws.copy()
    if all_nws_tmp[pos] == all_nws_sorted[0]:
        return 5
    elif all_nws_tmp[pos] == all_nws_sorted[1]:
        return 4
    elif all_nws_tmp[pos] == all_nws_sorted[2]:
        return 3
    elif all_nws_tmp[pos] == all_nws_sorted[3]:
        return 2
    else:
        return 1
```

```python
for index, row in dfNew.iterrows():
    dfNew.at[index, 'player_0_supposed_pos'] = get_supposed_position_1(row, 0)
    dfNew.at[index, 'player_1_supposed_pos'] = get_supposed_position_1(row, 1)
    dfNew.at[index, 'player_2_supposed_pos'] = get_supposed_position_1(row, 2)
    dfNew.at[index, 'player_3_supposed_pos'] = get_supposed_position_1(row, 3)
    dfNew.at[index, 'player_4_supposed_pos'] = get_supposed_position_1(row, 4)
    dfNew.at[index, 'player_5_supposed_pos'] = get_supposed_position_2(row, 0)
    dfNew.at[index, 'player_6_supposed_pos'] = get_supposed_position_2(row, 1)
    dfNew.at[index, 'player_7_supposed_pos'] = get_supposed_position_2(row, 2)
    dfNew.at[index, 'player_8_supposed_pos'] = get_supposed_position_2(row, 3)
    dfNew.at[index, 'player_9_supposed_pos'] = get_supposed_position_2(row, 4)
```

```python
def set_player(row, index, player_pos, i):

    player_str = 'player_' + str(player_pos)
    player = json.loads(row[player_str])
    dfPlayers.at[dfPlayers.loc[dfPlayers['account_id'] == player['account_id']].index[0], 'games'] += 1
    
    if row['radiant_win'] == True:
        if player_pos <= 4:
            dfPlayers.at[dfPlayers.loc[dfPlayers['account_id'] == player['account_id']].index[0], 'wins'] += 1
    else:
        if player_pos >= 5:
            dfPlayers.at[dfPlayers.loc[dfPlayers['account_id'] == player['account_id']].index[0], 'wins'] += 1

    
    supp_str = player_str + '_supposed_pos'
    if row[supp_str] == 1:
        dfPlayers.at[dfPlayers.loc[dfPlayers['account_id'] == player['account_id']].index[0], 'pos_1_num'] += 1
    elif row[supp_str] == 2:
        dfPlayers.at[dfPlayers.loc[dfPlayers['account_id'] == player['account_id']].index[0], 'pos_2_num'] += 1
    elif row[supp_str] == 3:
        dfPlayers.at[dfPlayers.loc[dfPlayers['account_id'] == player['account_id']].index[0], 'pos_3_num'] += 1
    elif row[supp_str] == 4:
        dfPlayers.at[dfPlayers.loc[dfPlayers['account_id'] == player['account_id']].index[0], 'pos_4_num'] += 1
    elif row[supp_str] == 5:
        dfPlayers.at[dfPlayers.loc[dfPlayers['account_id'] == player['account_id']].index[0], 'pos_5_num'] += 1
        

    dfPlayers.at[dfPlayers.loc[dfPlayers['account_id'] == player['account_id']].index[0], 'last_hits'] += player['last_hits']
    dfPlayers.at[dfPlayers.loc[dfPlayers['account_id'] == player['account_id']].index[0], 'denies'] += player['denies']
    dfPlayers.at[dfPlayers.loc[dfPlayers['account_id'] == player['account_id']].index[0], 'kills'] += player['kills']
    dfPlayers.at[dfPlayers.loc[dfPlayers['account_id'] == player['account_id']].index[0], 'deaths'] += player['deaths']
    dfPlayers.at[dfPlayers.loc[dfPlayers['account_id'] == player['account_id']].index[0], 'assists'] += player['assists']

    if i % 1000 == 0 and player_pos == 0:
        end = time.time()
        print(end - start, "seconds passed. i == ", i)
```

```python
dfPlayers = pd.DataFrame(columns=['account_id', 'games', 'wins', 'pos_1_num', 'pos_2_num', 'pos_3_num', 'pos_4_num', 'pos_5_num', 'last_hits', 'denies', 'kills', 'deaths', 'assists'])
dfPlayers['account_id'] = unique_ids
dfPlayers['games'] = 0
dfPlayers['wins'] = 0
dfPlayers['pos_1_num'] = 0
dfPlayers['pos_2_num'] = 0
dfPlayers['pos_3_num'] = 0
dfPlayers['pos_4_num'] = 0
dfPlayers['pos_5_num'] = 0
dfPlayers['last_hits'] = 0
dfPlayers['denies'] = 0
dfPlayers['kills'] = 0
dfPlayers['deaths'] = 0
dfPlayers['assists'] = 0
```

```python
start = time.time()
i = 0
for index, row in dfNew.iterrows():
    set_player(row, index, 0, i)
    set_player(row, index, 1, i)
    set_player(row, index, 2, i)
    set_player(row, index, 3, i)
    set_player(row, index, 4, i)
    set_player(row, index, 5, i)
    set_player(row, index, 6, i)
    set_player(row, index, 7, i)
    set_player(row, index, 8, i)
    set_player(row, index, 9, i)
    i += 1
    end = time.time()
```
